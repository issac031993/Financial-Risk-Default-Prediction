```python
stock_returns = np.log(market_risk_data.drop(['Date','dates'],axis=1)).diff(axis = 0, periods = 1) 
```

```python
stock_returns.shape
```

```python
stock_returns.head()
```

```python
stock_means = stock_returns.mean(axis = 0)
stock_means
```

```python
stock_sd = stock_returns.std(axis = 0)
stock_sd
```

```python
df = pd.DataFrame({'Average':stock_means, 'Volatility': stock_sd})
df
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Volatility'], df['Average'], color='blue', alpha=0.5)

# Add labels and title
plt.title('Stock Means vs Standard Deviation')
plt.xlabel('Standard Deviation (Volatility)')
plt.ylabel('Mean (Average)')

# Add a vertical and horizontal line for reference
plt.axvline(x=stock_sd.mean(), linestyle='--', color='red', label='Average Volatility')
plt.axhline(y=stock_means.mean(), linestyle='--', color='green', label='Average Mean')

# Add legend
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
